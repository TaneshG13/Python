# Arithmetic and math

### Arithmetic operators

```python
# addition
friends = 0
friends = friends + 1
friends += 1
print(friends)
```

```python
# substraction
friends -= 2
print(friends)
```

```python
# multiplication
friends = 5
friends *= 3
print(friends)
```

```python
# division
friends /= 3
print(friends)
```

```python
# exponetial
friends **= 2
print(friends)
```

```python
# modulas
friends %= 3
print(friends)
```

```python
# floor division
x = 14
y = 3
print(x//y)
```

### Math functions

```python
x = 3.14
y = -4
z = 5
```

```python
# rounding off
res = round(x)
print(res)
```

```python
# absolute value function
res = abs(y)
print(res)
```

```python
# power function
res = pow(z,3)
print(res)
```

```python
# max function
max = max(x,y,z)
print(max)
```

```python
# min function
min= min(x,y,z)
print(min)
```
