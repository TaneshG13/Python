# Typecasting

<b> The process of converting a variable from one data type to another datatype </b>

```python
name = "Jhon walker"
age = 21
gpa  = 3.01
is_student = True
```

```python
print(type(name))
```

```python
print(type(age))
```

```python
print(type(gpa))
```

```python
print(type(is_student))
```

```python
gpa = int(gpa)
print(gpa)
print(type(gpa))
```

```python
age = float(age)
print(age)
print(type(age))
```

```python
gpa = str(gpa)
print(gpa)
print(type(gpa))
```

```python
name = bool(name)
print(name)
print(type(name))
```

#### Note: False when string is empty otherwise True

```python
is_student = str(is_student)
print(is_student)
print(type(is_student))
```
