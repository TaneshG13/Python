# Python weight converter

```python
weight = float(input("Enter your weight: "))
unit = input("Kgs or lbs: ")
if unit == 'Kgs':
    weight_in_lbs = weight * 2.205
    print(f"Your weight in lbs is: {weight_in_lbs}")
elif unit == 'lbs':
    weight_in_Kgs = weight / 2.205
    print(f"Your weight in Kgs is: {weight_in_Kgs}")
else:
    print("Something went wrong!")
```
