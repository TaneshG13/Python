# Python Calculator

```python
operator = input("Enter an operator (+, -, *, /): ")
num1 = float(input("Enter the 1st number: "))
num2 = float(input("Enter the 2nd number: "))
res = None
if operator == '+':
    res = num1 + num2
elif operator == '-':
    res = num1 - num2
elif operator == '*':
    res = num1 * num2
elif operator == '/':
    res = num1 / num2
else:
    print("Something went wrong!")
print(f"Your result is: {res}")
```
