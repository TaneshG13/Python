# String Methods

```python
name = input("Enter Your full name: ")
```

```python
# len
res = len(name)
print(res)
```

```python
# find
res = name.find('n')
print(res)
```

Note: it gives index of first occurence

```python
# rfind
res = name.rfind('a')
print(res)
```

Note: it gives index of last occurence

```python
# capitalize
res = name.capitalize()
print(res)
```

Note: It Capitalizes first character of the string

```python
# Upper
res = name.upper()
print(res)
```

```python
# lower
res = name.lower()
print(res)
```

```python
# isdigit
res = name.isdigit()
print(res)
```

Note: Prints true if string contians digit

```python
# alpha
res = name.isalpha()
print(res)
```

Note: It returns true only if striing contains only alphabets. In this  case string contains whitespace hence Returns False 

```python
# count
phoneno = "122-123-122-1"
res = phoneno.count('-')
print(res)
```

```python
# replace
res = phoneno.replace('-',"")
print(res)
```
