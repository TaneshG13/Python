<b>
validate User input<br>
1. Username is no more than 12 characters<br>
2. Username must not contain spaces<br>
3. Username must not contain digit<b>

```python
username = input("Enter a username: ")
if len(username) <= 12 and len(username) > 0 and username.count(" ") == 0 and username.isdigit() == False:
    print("Valid")
else:
    print("Invalid")
```
