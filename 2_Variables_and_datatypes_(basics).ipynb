# Variables

<b>a container for a value (string, integer, float, boolean). a variable behaves as if it was the value it contains</B>

```python
#strings
first_name = "Jhon"
print(first_name)
```

Print using f strings (formatable strings)

```python
print(f"Hello! {first_name}")
```

```python
#integer
age = 21
print(f"Your are {age} years old.")
```

```python
#float
weight = 68.5
print(f"Your weight is {weight} kg")
```

```python
#boolean
is_student = True
print(f"Are you a student: {is_student}")
```
