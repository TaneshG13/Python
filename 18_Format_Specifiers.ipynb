# Format Specifiers

<b>{value:flags} a value based on what flags are inserted <\b>

.(number)f = round to that many decimal places (fixed point)<br>
:(number) = allocate that many spaces<br>
:03 = allocaate and zero pad that many spaces<br>
:< = left justify<br>
:> = right justify<br>
:^ = center align<br>
:+ = use a plus sign to indicate positive value<br>
:= = place sign to indicate positive value<br>
: = insert a space beefore positive numbers<br>
:, = comma separatoe<br>

```python
price1 = 3.14159
price2 = -987.65
price3 = 12.34
```

```python
print(f"price 1 is {price1}")
print(f"price 2 is {price2}")
print(f"price 3 is {price3}")
```

```python
print(f"price 1 is {price1:.2f}")
print(f"price 2 is {price2:.2f}")
print(f"price 3 is {price3:.2f}")
```

```python
print(f"price 1 is ${price1:10}")
print(f"price 2 is ${price2:10}")
print(f"price 3 is ${price3:10}")
```

```python
print(f"price 1 is ${price1:010}")
print(f"price 2 is ${price2:010}")
print(f"price 3 is ${price3:010}")
```

```python
print(f"price 1 is ${price1:<10}")
print(f"price 2 is ${price2:<10}")
print(f"price 3 is ${price3:<10}")
```

```python
print(f"price 1 is ${price1:>10}")
print(f"price 2 is ${price2:>10}")
print(f"price 3 is ${price3:>10}")
```

```python
print(f"price 1 is ${price1:^10}")
print(f"price 2 is ${price2:^10}")
print(f"price 3 is ${price3:^10}")
```

```python
print(f"price 1 is ${price1:+}")
print(f"price 2 is ${price2:+}")
print(f"price 3 is ${price3:+}")
```

```python
price1 = 3000.14159
price2 = -9870.65
price3 = 1200.34

print(f"price 1 is ${price1:,}")
print(f"price 2 is ${price2:,}")
print(f"price 3 is ${price3:,}")
```

```python
print(f"price 1 is ${price1:=}")
print(f"price 2 is ${price2:=}")
print(f"price 3 is ${price3:=}")
```

```python
print(f"price 1 is ${price1:}")
print(f"price 2 is ${price2:}")
print(f"price 3 is ${price3:}")
```
