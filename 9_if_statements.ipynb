# Conditional operators

## if Statements

<b>Do some code if some codition is true else do something else </b>

```python
age = int(input("Enter Your age: "))

if age >= 18:
    print("You are now signed up!")
```

```python
age = int(input("Enter Your age: "))

if age >= 18:
    print("You are now signed up!")
```

```python
# use of else
age = int(input("Enter Your age: "))

if age >= 18:
    print("You are now signed up!")
else:
    print("You must be 18+ to sign up.")
```

```python
# use of elif
age = int(input("Enter Your age: "))

if age >= 18:
    print("You are now signed up!")
elif age <= 0:
    print("You haven't been born yet!")
else:
    print("You must be 18+ to sign up.")
```

```python
age = int(input("Enter Your age: "))

if age >= 18 and age < 100:
    print("You are now signed up!")
elif age <= 0:
    print("You haven't been born yet!")
elif age >= 100:
    print("Are you still alive?")
else:
    print("You must be 18+ to sign up.")
```
