# Conditional Expressions

<b>A one line shortcut for if-else statement (known as ternary oprator in other programming languages)</b><br>
<b>Print or assign one of two based on a condition</b><br>
<b>X if condition = True else Y</b><br>

```python
num = 5
print("Positive" if num > 0 else "Negetive")
```

```python
a = 6
b = 7
max_num = a if a > b else b
print(max_num)
```
