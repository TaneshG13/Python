# Numbers

Python provides various ways to work with numbers, including integers, floating-point numbers, and complex numbers.

## Basic Arithmetic Operations

```python
# Addition
sum_result = 5 + 3
print(sum_result)
```

```python
# Subtraction
difference = 7 - 4
print(difference)
```

```python
# Multiplication
product = 2 * 6
print(product)
```

```python
# Division
quotient = 10 / 2
print(quotient)
```

## Exponentiation

```python
# Exponentiation
power_result = 2 ** 3
print(power_result)
```

## Working with Floats

```python
# Floating-point numbers
float_num = 3.14
```

```python
# Rounding
rounded_num = round(float_num, 2)
print(rounded_num)
```

```python
# Conversion from int to float
int_to_float = float(5)
print(int_to_float)
```
