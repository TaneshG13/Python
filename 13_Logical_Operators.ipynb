# Logical Operators

<b>evaluate multiple conditions</b><br>
<b>or = at least one condition must be true</b><br>
<b>and = both conditions must be true</b><br>
<b>not = inverts the condition (not Flase, not True)</b><br>

```python
# or
temp = 36
is_raining = False

if temp > 35 or temp < 0 or is_raining:
    print("The outdoor event is cancelled")
else:
    print("The Outdoor event will happen")
```

```python
# and 
temp = 36
is_sunny = True
if temp >= 28 and is_sunny:
    print("It is hot outside")
    print("It is sunny")
else:
    print("It is not hot outside")
```

```python
# not
is_sunny = True

if not is_sunny:
    print("You cant play outside!")
else:
    print("You can play outside!")
```
