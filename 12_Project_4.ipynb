# Python Temprature Converter

```python
unit = input("plese Enter temprature unit (C/ F): ")
temp = float(input("plese Enter temprature: "))
if unit == 'C':
    temp = (temp * (9/5)) + 32
    print(f"Temprature in F is: {temp}")
elif unit == 'F':
    temp = (temp - 32) * (5 / 9)
    print(f"Temprature in C is: {temp}")
else:
    print("Something went wrong!")
```
